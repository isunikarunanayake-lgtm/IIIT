## FOR LOOP

### FOR starts from 0 ends one before

```python
for i in range(5):
        print(i)
```

### step mean +2

```python
for i in range (10,36,2):
    print(i)
```

```python
for i in range (-100,10,5):
    print(i)
```

### Looping through list

```python
fruits_list=["apple","banana","cherry"]
for fruits in fruits_list:
    print(fruits)
```

### Looping through a dictionary

```python
student={"name":"isuni","age":"30","grade":"A"}
for key in student:
    print(key,":",student[key])
```

### Listing the letters

```python
for letter in "Python":
    print(letter)
```

### Keeping spaces

```python
for letter in "Python":
    print(letter,end="  ")
```

# while Loop

```python
i=0
while i<5:
    print(i)
    i+=1
```

## Repeat until conditions met

```python
password=""
while password !="python123":
    password=input("Enter the password")
print("Access Granted")
```

## waiting condition to change
#### This is fun in time.sleep(2) means that next battery level is showing up after 2 seconds.
   #### ex: battery level 10 show up after 2 sec battery level 9 show up

```python
import time
battery=10
while battery>0:
    print("Battery level:",battery)
    battery-=1
    time.sleep(2)
    print("Battery empty!")
```

## Validating user input

```python
age=-1
while age<0:
    age=int(input("Enter a valid age:"))
    if age>0:
        print("Thank you")
```
