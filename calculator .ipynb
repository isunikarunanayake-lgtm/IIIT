# Developing a Calculator

```python
print("*****Calculator******")
# menu
print("1.Addition")
print("2.Substraction")
print("3.Multiplication")
print("4.Division")
print("5.Exit")
print("-----------------")
choice=input("Enter the choice")
if choice=="5":
    print("Good Bye")
else:
    num1=int(input("Enter the the first number"))
    num2=int(input("Enter the the second number"))
    if choice=="1":
        answer=num1+num2
    elif choice=="2":
        answer=num1-num2
    elif choice=="3":
        answer=num1*num2
    elif choice=="4":
        answer=num1/num2
    print(answer)
```

# with Functions

```python
print("*****Calculator******")
# menu
print("1.Addition")
print("2.Substraction")
print("3.Multiplication")
print("4.Division")
print("5.Exit")
print("-----------------")

choice=input("Enter the choice")
if choice=="5":
    print("Good Bye")
else:
    def calculate(num1,num2):
         if choice=="1":
             return(num1+num2)
         elif choice=="2":
             return(num1-num2)
         elif choice=="3":
             return(num1*num2)
         elif choice=="4":
             return(num1/num2)

```

```python
calculate(23,45)
```
