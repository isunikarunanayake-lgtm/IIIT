# Create a empty list method01


```python
# Create a empty list method01
numbers=[]
print(numbers)
print(type(numbers))
```

```python
# Create a empty list method02
numbers=list()
print(numbers)
print(type(numbers))
```

```python
# Create a list with values method01
mylists=[11,23,34.44]
print(mylists)
print(type(mylists))
```

```python
#print numbers from 0 to 4
list=list(range(5))
print(list)
```

```python
# Create a list with values method02
mylists1=[12,34,56,455]
print(mylists1)
print(type(mylists1))

# Append Method
mylists1.append(89)
print(mylists1)

# Insert Method
mylists1.insert(1,77)
print(mylists1)
```

```python
# Access the value using index
mylist=[20,"john",34,'False',"Anne"]
print(mylist[0])
print(mylist[::-1])
print(mylist[:3])
print(mylist[:])
print(mylist[2:5])
print(mylist[2:])
print(mylist[2:5:2])
```

```python
# Nested Lists
mylist=[20,"john",[20,24,29,33]]
print(mylist[2])
print(mylist[2][1])
```

```python
# Removing items
mylist=[20,34,333,445]
print(mylist)
mylist.remove(20)
print(mylist)
mylist.pop()
print(mylist)
mylist.pop(1)
print(mylist)
mylist.clear()
print(mylist)
```

```python
# Other use
mylist=[45,78,666,89,56,43,49]
print(len(mylist))
print(mylist.index(45))
print(mylist.count(56))
```

```python
# Sorting and reversing
numbers=[5,6,7,3,4]
numbers.sort()
print(numbers)
numbers.reverse()
print(numbers)
```
