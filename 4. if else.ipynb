## if else clause

```python
age=60
if age>18:
    print("young")
else:
    print("adult")
```

```python
marks=int(input("Enter the marks"))
if 0<marks<100 and marks>=90:
    print("A")
elif 80<=marks<90:
    print("B")
elif 70<=marks<80:
    print("C")
elif 60<=marks<70:
    print("D")
elif marks<=60:
    print("F")
else:
    print("Invalid score Entered")
```

```python
marks=70
if marks>50:
    print("pass")
else:
    print("fail")
```

```python
marks=int(input("Enter the marks?"))
if marks>=75:
    print("Grade A")
elif marks>=65:
    print("Grade B")
elif marks>=50:
    print("Grade C")
else:
    print("Grade F")
          
```

```python
marks=int(input("Enter the Marks"))
if marks>=75:
    print("Grade A")
else:
    if marks>=65:
        print("Grade B")
    else:
        if marks>=50:
            print("Grade C")
        else:
            print("Grade F")
```

```python
num=int(input("Enter the num"))
if num%2==0:
    print("Even Number")
else:
    print("Odd Number")
```

```python
age=int(input("Enter the age"))
if age>=18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```
